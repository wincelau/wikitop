```python
import json
from mdutils.mdutils import MdUtils
from slugify import slugify

with open('../data/FR_38.geojson') as json_file:
    data = json.load(json_file)
    for fe in data['features']:
        p = fe['properties']
        
        osmid = fe.get('id').replace("way/", "").replace("node/", "")
        filename = "../lieux/"+osmid+".md"
        
        type = ""
        if p.get('leisure') == "sports_centre":
            type = "Artificiel"
        else:
            type = "Naturel"
            
        if p.get('natural'):
            type = "Naturel"
        
        categories = []
        
        
        if p.get('climbing:boulder'):
            categories.append("Bloc")
            
        if p.get('climbing:sport'):
            categories.append("Voie équipée")
            
        if p.get('climbing:trad'):
            categories.append("Terrain d'aventure")
            
        if p.get('climbing:toprope'):
            categories.append("Moulinette")
            
        if p.get('climbing:multipitch'):
            categories.append("Grande voie")
            
        if p.get('climbing:deepwater'):
            categories.append("Deep water")
            
        if p.get('climbing:ice'):
            categories.append("Cascade de glace")
            
        if p.get('climbing:mixed'):
            categories.append("Mixte")
            
            
        coordinates = None
        
        if fe['geometry']['type'] == 'Point':
            coordinates = fe['geometry']['coordinates']
        elif fe['geometry']['type'] == 'LineString':
            coordinates = fe['geometry']['coordinates'][0]
        elif fe['geometry']['type'] == 'Polygon':
            coordinates = fe['geometry']['coordinates'][0][0]
        
        f = open(filename, "w")
        f.write("---\n")
        f.write("layout: lieu\n")
        f.write("title: " + p.get('name', "")+"\n")
        f.write("type: "+type+"\n")
        f.write("categories:")
        if not len(categories):
            f.write(" []")
        f.write("\n")
        for ck in categories:
            f.write(" - "+ck+"\n")
        f.write("permalink: /lieux/"+slugify(p.get('name', ""))+"-"+osmid+"\n")
        f.write("coordinates: "+json.dumps(coordinates)+"\n")
        f.write("geometry: "+json.dumps(fe['geometry'])+"\n")
        f.write("osm: "+json.dumps(p)+"\n")
        f.write("---\n")
        f.close()
        
```
